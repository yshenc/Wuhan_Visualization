```python
import math
for i,d in data.iterrows():
    dic = {}
    dic['cat'] = data.loc[i,'省份']
    dic['value'] = math.log(data.loc[i,'确诊人数'],2)+1
    dic['name'] = data.loc[i,'省份']
    dic['icon'] = './img/'+ data.loc[i,'省份']+'.jpg'
    dic['desc'] = '确诊人数:  '+str(data.loc[i,'确诊人数']) + '<br><br>' +'治愈人数:  '+str(data.loc[i,'治愈人数']) + '<br><br>''死亡人数:  '+str(data.loc[i,'死亡人数'])
    result.append(dic)
```
