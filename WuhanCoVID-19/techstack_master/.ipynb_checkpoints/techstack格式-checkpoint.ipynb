```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('./疫情数据.xlsx')
```

```python
data
```

```python
result = []
```

```python
for i,d in data.iterrows():
    dic = {}
    dic['cat'] = data.loc[i,'省份']
    dic['name'] = data.loc[i,'城市']
    if data.loc[i,'省份'] == '湖北省':
        dic['value'] = data.loc[i,'确诊人数']/2
    else:
        dic['value'] = data.loc[i,'确诊人数']
    dic['icon'] = ''
    dic['desc'] = '确诊人数:'+str(data.loc[i,'确诊人数']) + '<br>' + '疑似人数:'+str(data.loc[i,'疑似人数']) + '<br>' +'治愈人数:'+str(data.loc[i,'治愈人数']) + '<br>''死亡人数:'+str(data.loc[i,'死亡人数'])
    result.append(dic)
```

```python
result
```

```python
with open("techstack格式数据.txt") as f:
    f.writelines(result)
```
